```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from mpl_toolkits import mplot3d
```

```python
import math
```

```python
pi = math.pi
```

```python
#sine
x = np.arange(0, 3*pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sine of X')
plt.xlabel('x')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("sin(x)")
plt.grid(True, which = 'both')
plt.show()
```

```python
#cosine
x = np.arange(0, 3*pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.title('Cosine of X')
plt.xlabel('x')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', '\u03C0/2', '\u03C0', '3\u03C0/2', '2\u03C0', '5\u03C0/2', '3\u03C0'])#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("cos(x)")
plt.grid(True, which = 'both')
plt.show()
```

```python
#sin(x) + cos(x)
x = np.arange(0, 3*pi, 0.1)
y = np.sin(x) + np.cos(x)
plt.plot(x, y)
plt.title('Sum of Sine and Cosine')
plt.xlabel('x')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', '\u03C0/2', '\u03C0', '3\u03C0/2', '2\u03C0', '5\u03C0/2', '3\u03C0'])#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("sin(x) + cos(x)")
plt.grid(True, which = 'both')
plt.show()
```

```python
#square of sin(x)
x = np.arange(0, 3*pi, 0.1)
y = np.square(np.sin(x))
plt.plot(x, y)
plt.title('Square of Sine of X')
plt.xlabel('x')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', '\u03C0/2', '\u03C0', '3\u03C0/2', '2\u03C0', '5\u03C0/2', '3\u03C0'])#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("sin^2(x)")
plt.grid(True, which = 'both')
plt.show()
```

```python
#square of cos(x)
x = np.arange(0, 3*pi, 0.1)
y = np.square(np.cos(x))
plt.plot(x, y)
plt.title('Square of Cosine of X')
plt.xlabel('x')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', '\u03C0/2', '\u03C0', '3\u03C0/2', '2\u03C0', '5\u03C0/2', '3\u03C0'])
#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("cos^2(x)")
plt.grid(True, which = 'both')
plt.show()
```

```python
#sin(x)*cos(x)
x = np.arange(0, 3*pi, 0.1)
y = np.sin(x) * np.cos(x)
plt.plot(x, y)
plt.title('Sin(x)*Cos(x)')
plt.xlabel('x')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', '\u03C0/2', '\u03C0', '3\u03C0/2', '2\u03C0', '5\u03C0/2', '3\u03C0'])
plt.axhline(0, color = 'r')
plt.ylabel("sin(x)*cos(x)")
plt.grid(True, which = 'both')
plt.show()
```

```python
#sin^2(x) + cos^2(x)
x = np.arange(0, 3*pi, 0.1)
y = np.square(np.sin(x)) + np.square(np.cos(x))
plt.plot(x, y)
plt.title('(Sine of X)^2 + (Cosine of x)^2')
plt.xlabel('x')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', '\u03C0/2', '\u03C0', '3\u03C0/2', '2\u03C0', '5\u03C0/2', '3\u03C0'])
#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
#plt.axhline(0, color = 'r')
plt.ylabel("sin^2(x) + cos^2(x)")
plt.grid(True, which = 'both')
plt.show()
```

```python
#3D graph of complex number
ax = plt.axes(projection='3d')
x = np.arange(0, 10*pi, 0.1)
y = np.sin(x)
z = np.cos(x)
plt.title('Complex Number : sin(\u03A9*t) + i*cos(\u03A9*t)')
ax.set_xlabel('\u03A9' + '*t')
ax.set_ylabel('sin(' + '\u03A9' + '*t)')
ax.set_zlabel('cos(' + '\u03A9' + '*t)')
ax.plot3D(x, y, z)
```

```python
#f(x) = x*x + 2
x = np.arange(0, 10, 0.1)
y = [i*i + 2 for i in x]
plt.plot(x, y)
plt.title('f(x) = x*x + 2')
plt.xlabel('x')
plt.axhline(0, color = 'r')
#fig, ax = plt.subplots()
#plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
#plt.axhline(0, color = 'r')
plt.ylabel("y")
plt.grid(True, which = 'both')
plt.show()
```

```python
#f(x) = x*x - 8*x + 6
x = np.arange(0, 10, 0.1)
y = [i*i - 8*i + 6 for i in x]
plt.plot(x, y)
plt.title('f(x) = x*x - 2*x')
plt.xlabel('x')
plt.axhline(0, color = 'r')
#fig, ax = plt.subplots()
#plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
#plt.axhline(0, color = 'r')
plt.ylabel("y")
plt.grid(True, which = 'both')
plt.show()
```

```python
#sum of two functions
x = np.arange(0, 10, 0.1)
y = [i*i - 8*i + 6 + i*i + 2 for i in x]
plt.plot(x, y)
plt.title('f(x) = x*x - 2*x')
plt.xlabel('x')
plt.axhline(0, color = 'r')
#fig, ax = plt.subplots()
#plt.xticks([0, pi/2, pi, 3*(pi/2), 2*pi, 5*(pi/2), 3*pi], ['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
#plt.axhline(0, color = 'r')
plt.ylabel("y")
plt.grid(True, which = 'both')
plt.show()
```

```python
#Plot of sinc(x): Cardic Function
x = np.arange(-2*pi, 2*pi, 0.1)
y = np.sin(pi*x)/(pi*x)
plt.plot(x, y)
plt.title('Sinc of X')
plt.xlabel('x')
#fig, ax = plt.subplots()
plt.xticks([-2*pi ,-3*pi/2, -pi, -pi/2, 0, (pi/2), pi, 3*(pi/2), 2*pi], ['-2\u03C0' ,'-3\u03C0/2', '-\u03C0', '-\u03C0/2', '0', '\u03C0/2', '\u03C0', '3\u03C0/2', '2\u03C0'])
#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("sin(x)")
plt.grid(True, which = 'both')
plt.show()
```
