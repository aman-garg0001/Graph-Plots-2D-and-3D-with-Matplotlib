```python

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import math
pi = math.pi
```

```python
cf = 25   # carrier frequency
mf = 1     # message frequency
sf = 10000 # sampling frequency
df = 15     # frequency response

t = np.arange(0,2*pi,1/sf)   # time domain
```

```python
# Message signal
mt = np.cos((2*pi*mf)*t)
plt.figure(figsize=(30,5))
plt.plot(t,mt)
plt.title("Message Signal")
plt.xlabel('t')
plt.ylabel('m(t)')
plt.xticks([0,pi/2,pi,3*pi/2,2*pi],['0', '\u03C0/2', '\u03C0','1.5\u03C0','2\u03C0'])
plt.axhline(0,color='r')
plt.grid(True)

plt.show()
```

```python
# Carrier signal
ct = np.cos((2*pi*cf)*t)
plt.figure(figsize=(30,5))
plt.plot(t,ct)
plt.title("Carrier Signal")
plt.xlabel('t')
plt.ylabel('c(t)')
plt.xticks([0,pi/2,pi,3*pi/2,2*pi],['0', '\u03C0/2', '\u03C0','1.5\u03C0','2\u03C0'])
plt.axhline(0,color='r')
plt.grid(True)

plt.show()
```

```python
# Frequency Modulated Signal
fmt = np.cos((2*pi*cf)*t + df*mt)
plt.figure(figsize=(30,5))
plt.plot(t,fmt)
plt.title("Frequency Modulated Signal")
plt.xlabel('t')
plt.ylabel('fm(t)')
plt.xticks([0,pi/2,pi,3*pi/2,2*pi],['0', '\u03C0/2', '\u03C0','1.5\u03C0','2\u03C0'])
plt.axhline(0,color='r')
plt.grid(True)

plt.show()
```
