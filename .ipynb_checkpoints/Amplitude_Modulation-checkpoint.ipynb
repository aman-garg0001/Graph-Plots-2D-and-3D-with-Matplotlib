```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import math
pi = math.pi
```

```python
carrier_frequency = 20
message_frequency = 2
```

```python
# Message Signal
t = np.arange(0, pi, 0.00001)
message_signal = np.cos(2*pi*message_frequency*t)
plt.figure(figsize=(15,5))
plt.plot(t, message_signal)
plt.title('Message Signal')
plt.xlabel('t')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi], ['0', '\u03C0/2', '\u03C0'])#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("cos(\u03A9c*t)")
plt.grid(True, which = 'both')
plt.show()
```

```python
# Carrier Signal\
t = np.arange(0, pi, 0.00001)
plt.figure(figsize=(15,5))
carrier_signal = np.cos(2*pi*carrier_frequency*t)
plt.plot(t, carrier_signal)
plt.title('Carrier Signal')
plt.xlabel('t')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi], ['0', '\u03C0/2', '\u03C0'])#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("cos(\u03A9m*t)")
plt.grid(True, which = 'both')
plt.show()
```

```python
#Amplitude Modulation
x = np.arange(0, pi, 0.00001)
y = np.cos(2*pi*message_frequency*x)*np.cos(2*pi*carrier_frequency*x)
plt.figure(figsize=(15,5))
plt.plot(x, y)
plt.title('AM Signal')
plt.xlabel('t')
#fig, ax = plt.subplots()
plt.xticks([0, pi/2, pi], ['0', '\u03C0/2', '\u03C0'])#ax.set_xticklabels(['0', 'pi/2', 'pi', '3pi/2', '2pi', '5pi/2', '3pi'])
plt.axhline(0, color = 'r')
plt.ylabel("cos(\u03A9m*t)*cos(\u03A9c*t)")
plt.grid(True, which = 'both')
plt.show()
```
